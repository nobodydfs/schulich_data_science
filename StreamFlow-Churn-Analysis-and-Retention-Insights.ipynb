From above, 246 individuals have already cancelled subscriptions and 754 individuals have not cancelled yet. That's why there are 754 missing values for both 'Date_of_Churn' and 'Reason for Churn'

##### Churn Rate

```python
#Calculate the churn rate 
#a binary varible 'churned' where 1 means churned and 0 means not churned
churned_subscribers = df['Churned'].sum()
total_subscribers = len(df)
churn_rate = churned_subscribers/total_subscribers
churn_rate
```

#### Monthly Recurring Revenue (mrr)

```python
subscribers = df[df['Churned'] == 0]
mrr = subscribers['Monthly_Payment'].sum()
mrr
```

#### Average Revenue Per User (ARPU)

```python
arpu = mrr /subscribers['User_ID'].nunique()
arpu
```

##### Customer Lifetime Value (CLTV)
##### Assuming CLTV as ARPU multiplied by average subscription duration for churners

```python
churner = df[df['Churned'] == 1]
churner['Subscription_Duration'] = (churner['Date_of_Churn'] - churner['Subscription_Start_Date']).dt.days / 30
average_subscription_duration = churner['Subscription_Duration'].mean()
cltv = arpu * average_subscription_duration
cltv
```

### BUSINESS IMPACT ESTIMATE

#### Quantitative Benefits

```python
users = 10000 # with the base of 10000 users
new_churn_rate = churn_rate * (1- 0.05) # reduce churn rate by 5%
new_churn_rate # new churn rate is 0.2337
```

```python
churned_users = churn_rate * users # churned users among 10000 users
churned_users
```

```python
reduced_churned_users = new_churn_rate * users# churned userd after reducing 5% churn rate 
reduced_churned_users
```

```python
#addtional retained users per month
additional_users_retained = churned_users - reduced_churned_users
additional_users_retained
```

```python
#addtional revenue gained per month
additional_revenue =additional_users_retained * arpu
additional_revenue
```
